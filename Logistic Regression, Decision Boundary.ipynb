```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1]).reshape(-1,1)
```

```python
def sigmoid(z):
    res = 1/(1+np.exp(-z))

    return res
```

```python
z = np.arange(-10, 11)
fig, ax = plt.subplots(1, 1, figsize=(5, 3))
ax.plot(z, sigmoid(z), c='b')
ax.set_title("Sigmoid function")
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')
```

```python
x0 = np.arange(0, 6)
x1 = 3 - x0

fig, ax = plt.subplots(1, 1, figsize=(5, 4))

# Plot decision boundary
ax.plot(x0, x1, c="b", label="Decision boundary")
ax.fill_between(x0, x1.flatten(), alpha=0.2)

# Plot points as scatter, color-coded by class
ax.scatter(X[y[:,0] == 0][:,0], X[y[:,0] == 0][:,1], c='r', label='Class 0')
ax.scatter(X[y[:,0] == 1][:,0], X[y[:,0] == 1][:,1], c='g', label='Class 1')

ax.axis([0, 4, 0, 3.5])
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.legend()
ax.set_title("Binary Classification Dataset with Decision Boundary")
plt.grid(True)
plt.show()
```
