```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
m = x_train.shape[0]
n = len(x_train)
```

```python
print(f"length: {n}")
print(f"length: {m}")
```

```python
i=0
x_i = x_train[i]
y_i = y_train[i]
print(f"x: {x_i}, y: {y_i}")
```

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("Housing Prices")
plt.ylabel("Price")
plt.xlabel("Size")
plt.show()
```

```python
def compute_model(x, w, b):
    m = len(x)
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i] + b

    return f_wb
```

```python
tmp_f_wb = compute_model(x_train, 200, 100)
plt.plot(x_train, tmp_f_wb, c='b', label="Our Prediction") 
plt.scatter(x_train, y_train, marker="x", c="r", label="Actual Label")
plt.title("Housing Prices")
plt.xlabel("Size")
plt.ylabel("Price")
plt.legend()
plt.show()
```
